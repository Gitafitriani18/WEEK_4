```python
# ==============================
# ENCAPSULATION + INHERITANCE + POLYMORPHISM
# Tema: Kendaraan
# ==============================

# Parent Class
class Vehicle:
    def __init__(self, brand, speed):
        self._brand = brand
        self._speed = speed

    # Getter
    def get_speed(self):
        return self._speed

    # Setter
    def set_speed(self, speed):
        if speed > 0:
            self._speed = speed
        else:
            print("Kecepatan tidak valid!")

    # Method Polymorphism
    def move(self):
        print(self._brand, "bergerak")


# Child Class
class Car(Vehicle):
    def move(self):
        print(self._brand, "mobil melaju di jalan dengan kecepatan", self._speed, "km/jam")


class Motorcycle(Vehicle):
    def move(self):
        print(self._brand, "motor melaju lebih cepat dengan kecepatan", self._speed, "km/jam")


# Object
car1 = Car("Toyota", 120)
motor1 = Motorcycle("Honda", 100)

# Encapsulation
print("Kecepatan awal mobil:", car1.get_speed())

car1.set_speed(150)

print("Kecepatan setelah diubah:", car1.get_speed())

print()

# Polymorphism
car1.move()
motor1.move()
```
