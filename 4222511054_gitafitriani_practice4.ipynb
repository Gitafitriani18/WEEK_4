```python
class User:
    def __init__(self, username, email, gender):
        self._username = username
        self._email = email
        self._gender = gender

    # getter
    def get_username(self):
        return self._username

    def get_email(self):
        return self._email

    def get_gender(self):
        return self._gender

    # setter
    def set_username(self, username):
        self._username = username

    def set_email(self, email):
        self._email = email

    def set_gender(self, gender):
        self._gender = gender


# membuat object
user = User("Gita", "gita72@gmail.com", "female")

# menampilkan gender awal
print(user.get_gender())

# mengubah gender
user.set_gender("unknown")

# menampilkan gender setelah diubah
print(user.get_gender())
```
